```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation, rc
# from PIL import Image
```

```python
vec_dl = np.loadtxt("wavelength: 632.8 r0: 10 n_flt: 100 n_sc: 1000 n_R: 20 R_min: 300 R_max: 500__20191119_101756.txt")
vec_dl
```

```python
mat = vec_dl.reshape(20, 1000, 1000)
```

```python
fig = plt.figure()
ims = []
for i in range(20):
    im = plt.imshow(np.log(mat[i]))
    ims.append([im])
plt.show()
```

```python
#アニメーション表示の準備
print('making animation......')
ani = animation.ArtistAnimation(fig, ims, interval=100, blit=True)

#ここにアニメを表示する場合はこちらを有効に
rc('animation', html='jshtml')
ani
```

```python
ani.save('circle.gif', writer='imagemagick')
```
